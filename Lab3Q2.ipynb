```python
import pickle
import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from tabulate import tabulate
```

```python
f=open('X_train.pkl', 'rb')
x_train = pickle.load(f)

```

```python
f=open('Y_train.pkl', 'rb')
y_train = pickle.load(f)
```

```python
f=open('X_test.pkl', 'rb')
x_test = pickle.load(f)
```

```python
f=open('Fx_test.pkl', 'rb')
y_test = pickle.load(f)
x_test=x_test.reshape([80,1])
y_test=y_test.reshape([80,1])
```

```python
fig, axs = plt.subplots(20,figsize=(8,60))
for i in range(0,20):
    axs[i].scatter(x_test,y_test)
```

```python
train_1=[]
for i in range(0,20):
    train_1.append(x_train[i:(i+1)])
Xtrain= np.asarray(train_1)

train_2=[]
for i in range(0,20):
    train_2.append(y_train[i:(i+1)])
Ytrain= np.asarray(train_2)
Xtrain=Xtrain.reshape([20,400,1])
Ytrain=Ytrain.reshape([20,400,1])
Xtrain
```

```python
polynomial_feature = PolynomialFeatures(1)
new_testx= polynomial_feature.fit_transform(x_test)
y_predict = []
for i in range (0,20) :
    polynomial_feature = PolynomialFeatures(1)
    new_x = polynomial_feature.fit_transform(Xtrain[i])
    model= LinearRegression().fit(new_x, Ytrain[i])
    y_predict.append(model.predict(new_testx))
predicty= np.asarray(y_predict)
```

```python
fig, axs = plt.subplots(20,figsize=(8,60))
for i in range(0,20):
    axs[i].scatter(x_test,predicty[i])
    axs[i].scatter(x_test,y_test)
```

```python
plt.scatter(x_test,y_test)
plt.scatter(x_test,predicty[19]);
```

```python
bias=np.abs(np.mean(y_predict, axis=0)- y_test)
var= np.var(y_predict, axis=0)
bias1=np.mean(bias)
var1=np.mean(var)
```
